### 2. Handling Missing Values

<b>Parwise Deletion</b> is used when values are missing completely at random i.e MCAR. During Pairwise deletion, only the missing values are deleted. All operations in pandas like mean,sum etc intrinsically skip missing values.

<b>Listwise deletion</b>, complete rows(which contain the missing values) are deleted. As a result, it is also called Complete Case deletion. Like Pairwise deletion, listwise deletions are also only used for MCAR values.

The Listwise deletion problem is that a major chunk of data and hence a lot of information is lost. Hence, it is advisable to use it only when the number of missing values is very small.

<b>Dropping complete columns</b> If a column contains a lot of missing values, say more than 70%, and the feature is not significant for the task at hand, you might want to delete that feature. However, again, it is not a good methodology to delete data.

```python
data.Age.mean()
```

```python
data = data.dropna(subset=['RoomService'], axis=0)
```

```python
data.isnull().sum()
```

```python
#droping entire column because is contain more than 70% missing values
data = data.drop(columns='Cabin')
```

```python
data.isnull().sum()
```

```python
data
```

```python
#simpleImputer is used for handling missing values with specific 'constant', 
#fill_value is all missing values will be replaced with 0
constant_imputer = SimpleImputer(strategy = 'constant', fill_value = 0)
```

```python
data['FoodCourt'] = constant_imputer.fit_transform([data['FoodCourt']])[0] 
# we hypothesize that the missing values actually represent that people haven't spent any money on food
```

```python
data.isnull().sum()
```

```python
#strategy= most_frequent replaces missing values with the most frequently occuring value(mode)
frequent_imputer = SimpleImputer(strategy = 'most_frequent')
```

```python
data['HomePlanet'].value_counts().plot.bar()
plt.show()
```

```python
data['HomePlanet'] = data['HomePlanet'].astype(str) 
# Simple imputer with most_frequent works with numbers (integers) and strings
```

```python
data['HomePlanet'] = frequent_imputer.fit_transform(data['HomePlanet'].values.reshape(1, -1))[0]  
# should fill missing values with 'Earth' since it's the most frequent
```

```python
data.isnull().sum()
```

```python
data.Age.plot.hist()
plt.show()
```

```python
#should fill with mean of age
data.Age = data.Age.fillna(data.Age.mean())
```

```python
data.isnull().sum()
```

```python
data.ShoppingMall.plot.hist()
```

```python

data.ShoppingMall.mean()
```

```python
data.ShoppingMall.median()
```

```python
data.ShoppingMall = data.ShoppingMall.fillna(data.ShoppingMall.median())
```

```python
data.isnull().sum()
```

```python
data[['VRDeck', 'Spa']].plot.hist(subplots=True)
plt.show()
```

```python
scaler = MinMaxScaler() #min-max scales the values of vrder and spa between 0 and 1
                        #this is necessary because knn works best when data is normalized
scaled_data = scaler.fit_transform(data[['VRDeck', 'Spa']])
knn_imputer = KNNImputer(n_neighbors=5) # finds the 5 neares neighbors fro each missing value and iputes it with the average of those
imputed_data = knn_imputer.fit_transform(scaled_data)
```

```python
#reverses the min-max scaling, bringing the imputed values back to their origin scale
imputed_data=scaler.inverse_transform(imputed_data)
#this is necesasary because knn imputation was performed on the scaled data, and now we need to restore it
```

```python
imputed_data
```

```python
data[['VRDeck','Spa']]=imputed_data
```

```python
data.isnull().sum()
```

### Multivariate feature imputation - Multivariate imputation by chained equations (MICE)

A strategy for imputing missing values by modeling each feature with missing values as a function of other features in a round-robin fashion. It performns multiple regressions over random sample ofthe data, then takes the average ofthe multiple regression values and uses that value to impute the missing value. In sklearn, it is implemented as follows:

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(data[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].corr(),annot=True)
plt.show()
```

```python
#mice-> uses a regression model to predict missing values
mice=IterativeImputer()
imputed_values=mice.fit_transform(data[['VRDeck', 'FoodCourt', 'Spa']])
#iteratively predicts each missing value using the other available columns as predictors
data[['VRDeck', 'FoodCourt', 'Spa']]=imputed_values
```

```python
data.isnull().sum()
```

### 3. Time - Series Missing Value Imputation

You can use MICE and KNN for time series as well since they will use the dependencies between variables.

```python
aq_data = pd.read_csv('air_quality_missing.csv')
```

```python
aq_data.head(50)
```

For NO2 we have missing values at the start

```python

aq_data.isnull().sum()
```

```python

aq_data.tail(20)
```


For CO we have missing values at the end

```python
aq_data['NO2'] = aq_data['NO2'].bfill()
```

```python
aq_data.head()
```

```python

aq_data['CO'] = aq_data['CO'].ffill()
```

```python
aq_data.isnull().sum()
```

```python
aq_data['PM10'].plot.line()
plt.show()
```

```python
aq_data['PM10'] = aq_data['PM10'].interpolate(limit_direction='both')
```

```python
#label encoder converts categorical values into numerical labels 
encoder = LabelEncoder()
encoded_values = encoder.fit_transform(data['HomePlanet'])
```

```python
#convert boolean true/false into 1 and 0, leaves float values unchanged to avoid errors
def encode_bool_variable(x):
  if type(x) == float:
    return x
  return 1 if x else 0
```

```python
data['CryoSleep_Enc'] = data.CryoSleep.apply(encode_bool_variable)
```
