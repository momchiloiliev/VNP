**About the Dataset**  
This dataset focuses on air quality assessment across various regions. The dataset contains 5,000 samples and captures critical environmental and demographic factors that influence pollution levels.

**Features**:  
- **Temperature (°C)**: Average temperature of the region.  
- **Humidity (%)**: Relative humidity recorded in the region.  
- **PM2.5 Concentration (µg/m³)**: Levels of fine particulate matter.  
- **PM10 Concentration (µg/m³)**: Levels of coarse particulate matter.  
- **NO2 Concentration (ppb)**: Nitrogen dioxide levels.  
- **SO2 Concentration (ppb)**: Sulfur dioxide levels.  
- **CO Concentration (ppm)**: Carbon monoxide levels.  
- **Proximity to Industrial Areas (km)**: Distance to the nearest industrial zone.  
- **Population Density (people/km²)**: Number of people per square kilometer in the region.  

**Target Variable**: **Air Quality**  
- **Good**: Clean air with low pollution levels.  
- **Moderate**: Acceptable air quality but with some pollutants present.  
- **Poor**: Noticeable pollution that may cause health issues for sensitive groups.  
- **Hazardous**: Highly polluted air posing serious health risks to the population.  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import MinMaxScaler,StandardScaler,LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, r2_score
from keras.models import Sequential
from keras.layers import Dense, Input, Dropout
from xgboost import XGBClassifier
```

```python
data=pd.read_csv('pollution_dataset.csv')
```

```python
data.sample(10)
```

```python
data['Air Quality'].unique()
```

```python
def label_data(data:pd.DataFrame, columns:list):
  encoder = LabelEncoder()
  data_copy = data.copy()

  for column in columns:
    data_copy[column] = encoder.fit_transform(data_copy[[column]].astype(str).values.ravel())

  return data_copy
```

```python
data=label_data(data=data,columns=['Air Quality'])
```

```python
data.sample(3)
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(data.corr(),annot=True,fmt='.3f',linewidths=2)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['Air Quality'],axis=1)
Y = data['Air Quality']
Y = to_categorical(Y, num_classes=4)
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
model1 = Sequential([
    Input(shape=(X.shape[1],)),
    Dense(32, kernel_initializer="uniform", activation="relu"),
    Dense(16, kernel_initializer="uniform", activation="relu"),
    Dense(8, kernel_initializer="uniform", activation="relu"),
    Dense(4, kernel_initializer="uniform", activation="sigmoid"),
])
```

```python
model1.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
model1.summary()
```

```python
history = model1.fit(X_train, y_train, validation_split=0.1, epochs=50, batch_size=32)
```

```python
sns.lineplot(history.history['loss'], label='loss')
sns.lineplot(history.history['val_loss'], label='val_loss')
```

```python
y_pred = model1.predict(X_test).round()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

### MODEL 2

```python
model2 = Sequential([
    Input(shape=(X.shape[1],)),
    Dense(128, kernel_initializer="he_normal", activation="relu"),
    Dropout(0.3),
    Dense(64, kernel_initializer="he_normal", activation="relu"),
    Dropout(0.3),
    Dense(32, kernel_initializer="he_normal", activation="relu"),
    Dense(4, kernel_initializer="he_normal", activation="softmax"),
])
model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['f1_score'])
```

```python
model2.compile(
    loss="categorical_crossentropy",
    optimizer="adam",
    metrics=["f1_score"]
)
```

```python
history = model2.fit(X_train, y_train, validation_split=0.2, epochs=30, batch_size=32)
```

```python
sns.lineplot(history.history['loss'], label='loss')
sns.lineplot(history.history['val_loss'], label='val_loss')
```

```python
y_pred = model2.predict(X_test).round()
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix

accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
model3 = Sequential([
    Input(shape=(X.shape[1],)),
    Dense(32, kernel_initializer="uniform", activation="relu"),
    Dense(4, kernel_initializer="uniform", activation="softmax"),
])
model3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['f1_score'])
```

```python
history = model3.fit(X_train, y_train, validation_split=0.2, epochs=30, batch_size=32)
```

```python
y_pred = model3.predict(X_test).round()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
sns.lineplot(history.history['loss'], label='loss')
sns.lineplot(history.history['val_loss'], label='val_loss')
```

Моделот 2 е подобар од останатите бидејќи користи "He Normal" иницијализација на тежини која е посоодветна за ReLU активациски функции и овозможува побрзо конвергирање на тренингот. Дополнително, користи Dropout слоеви со стапка од 0.3 за да се спречи пренагласување (overfitting), што го прави поотпорен на некохерентност при работење со нови податоци. 
Во споредба со останатите модели, Моделот 1 има повеќе скриени слоеви, но користи "uniform" иницијализација која е помалку ефикасна за ReLU функциите, додека Моделот 3 е наједноставен со помал број на параметри и слоеви, што го прави побрз за тренирање, но послабо генерално за комплексни податоци.
