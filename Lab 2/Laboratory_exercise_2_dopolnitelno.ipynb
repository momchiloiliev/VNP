# Dataset Description

## Attributes

### People
* ID: Customer's unique identifier
* Year_Birth: Customer's birth year
* Education: Customer's education level
* Marital_Status: Customer's marital status
* Income: Customer's yearly household income
* Kidhome: Number of children in customer's household
* Teenhome: Number of teenagers in customer's household
* Dt_Customer: Date of customer's enrollment with the company
* Recency: Number of days since customer's last purchase
* Complain: 1 if the customer complained in the last 2 years, 0 otherwise

### Products
* MntWines: Amount spent on wine in last 2 years
* MntFruits: Amount spent on fruits in last 2 years

### Promotion
* NumDealsPurchases: Number of purchases made with a discount
* AcceptedCmp1: 1 if customer accepted the offer in the 1st campaign, 0 otherwise
* AcceptedCmp2: 1 if customer accepted the offer in the 2nd campaign, 0 otherwise

### Place
* NumWebPurchases: Number of purchases made through the company’s website
* NumCatalogPurchases: Number of purchases made using a catalogue
* NumStorePurchases: Number of purchases made directly in stores
* NumWebVisitsMonth: Number of visits to company’s website in the last month

### Target
The column Response is the prediction target.
Response: 1 if customer accepted the offer in the last campaign, 0 otherwise

## Handling the Missing Values
Handle the missing values using suitable method based on the insights obtained from the various visualizations.

```python
data_cleaned = data.copy()
```

```python
data_cleaned
```

```python
# Id is not crucial for analysis
data_cleaned.drop(columns=['ID'], inplace=True)
```

```python
# Columns with Significant Missing Data:

# Income: Given its importance, we could fill missing values using the median, as income often has outliers.
# Response: As this represents a target variable for many analyses, dropping rows with missing values in Response may be reasonable unless preserving them for specific analyses is essential.
# Columns with Moderate Missing Data:

# For numerical columns related to purchases (e.g., MntWines, NumWebPurchases, NumCatalogPurchases), using the median is also appropriate, as it mitigates the impact of outliers.
# Categorical columns (e.g., Education, Marital_Status): These could be filled with the most common value (mode) within each category, maintaining the distribution of responses.
# Columns with a Low Priority or Low Importance:

# ID: If customer IDs are not crucial for analysis, we can drop rows with missing IDs.
```

```python
# Filling missing values for 'Income' and other numerical columns with the median
data_cleaned['Income'] = data_cleaned['Income'].fillna(data_cleaned['Income'].median())
numerical_columns = ['MntWines', 'MntFruits', 'MntGoldProds', 'NumDealsPurchases', 
                     'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 
                     'Recency', 'Year_Birth', 'Kidhome', 'Teenhome', 'AcceptedCmp1', 'AcceptedCmp2']
for col in numerical_columns:
    data_cleaned[col] = data_cleaned[col].fillna(data_cleaned[col].mode()[0])
```

```python
data_cleaned.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
# Filling missing values for categorical columns with the mode (most common value)
categorical_columns = ['Education', 'Marital_Status']
for col in categorical_columns:
    data_cleaned[col] = encoder.fit_transform(data_cleaned[col])
```

```python
data_cleaned.isnull().sum()
```

```python
from datetime import datetime

# Convert the Dt_Customer column to datetime
data_cleaned['Dt_Customer'] = pd.to_datetime(data_cleaned['Dt_Customer'], errors='coerce', dayfirst=True)

# Calculate the number of years as a customer
today = datetime.today()
data_cleaned['Dt_Customer_new'] = (today - data_cleaned['Dt_Customer']).dt.days / 365.25

# Replace missing values in Dt_Customer_new with the average tenure
average_tenure = data_cleaned['Dt_Customer_new'].mean()
data_cleaned['Dt_Customer_new'] = data_cleaned['Dt_Customer_new'].fillna(average_tenure)

# Drop the original Dt_Customer column if it's no longer needed
data_cleaned.drop(columns=['Dt_Customer'], inplace=True)
```

```python
data_cleaned.isnull().sum()
```

```python
# Drop rows where 'Response' is NaN, as it's the target column
data_cleaned = data_cleaned.dropna(subset=['Response'])

# Ensure the 'Response' column contains only valid binary values (0 and 1)
valid_responses = [0, 1]
data_cleaned = data_cleaned[data_cleaned['Response'].isin(valid_responses)]

# Convert 'Response' column to integer type if needed
data_cleaned['Response'] = data_cleaned['Response'].astype(int)
```

```python
data_cleaned.isnull().sum()
```

## Dataset Splitting
Partition the dataset into training and testing sets with an 80:20 ratio.

```python
from sklearn.model_selection import train_test_split

X = data_cleaned.drop(columns=['Response'])
Y = data_cleaned['Response']
```

## Feature Preprocessing


Preprocess the features to make them ready for using the model

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
X_train,X_test,Y_train,Y_test
```

## K-Nearset Neighbors Classification

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN.fit(X_train, Y_train)
```

Use the trained model to make predictions for the test set.

```python
y_pred = KNN.predict(X_test)
```

```python
y_pred[:30]
```

Assess the performance of the model by using different classification metrics provided by the `scikit-learn` library.

```python
from sklearn.metrics import classification_report, accuracy_score

classification_report = classification_report(Y_test, y_pred)
accuracy = accuracy_score(Y_test, y_pred)

print(classification_report)
print(accuracy)
```
